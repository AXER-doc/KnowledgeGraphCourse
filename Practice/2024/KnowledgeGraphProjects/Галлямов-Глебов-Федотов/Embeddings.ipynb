###### Load graph and prepare DataFrame

```python
from rdflib import Graph, Literal, Namespace
import pandas as pd

graph = Graph()
graph.parse("final_tier_effect_fix.omn")

FGO = Namespace("http://kg@se.ifmo.ru/~s367583/fgo#")

subjects = []
predicates = []
objects = []

for subject, predicate, objekt in graph.triples((None, None, None)):
    if isinstance(objekt, Literal):
        continue
    try:
        subjects.append(graph.qname(subject))
        predicates.append(graph.qname(predicate))
        objects.append(graph.qname(objekt))
    except ValueError:
        print("not converted:", subject, predicate, objekt)

df = pd.DataFrame({
    "subject": subjects,
    "predicate": predicates,
    "object": objects
})
df.head()
```

```python
def extract_roles(servant_ids, color):
    quest = next(q for q in recs_v2.values() if QUEST in q["url"])
    roles = {
        inv_names[name]: svt["role"]
        for name, svt in quest["servants"].items()
        if svt["desc_color"] == color and name in inv_names
    }
    return roles


def get_suggestions(graph: Graph, quest_name: str, predicate: str) -> list[str]:
    """Get servant IDs suggested for a quest with given predicate."""
    FGO = Namespace("http://kg@se.ifmo.ru/~s367583/fgo#")
    return [graph.qname(obj) for obj in graph.objects(FGO[quest_name], FGO[predicate])]


def draw_suggestion_clusters(
    graph: Graph,
    quest_name: str,
    predicate: str,
    model,
    roles: dict,
    names: dict,
) -> None:
    """Draw clusters of suggested servants for a quest."""
    # Get servant IDs and their embeddings
    servant_ids = get_suggestions(graph, quest_name, predicate)
    servant_embeddings = dict(zip(servant_ids, model.get_embeddings(servant_ids)))

    # Prepare 2D embeddings
    # embeddings_2d = PCA(n_components=2).fit_transform(
    #     np.array([i for i in servant_embeddings.values()])
    # )
    embeddings_2d = TSNE().fit_transform(np.array(list(servant_embeddings.values())))

    # Cluster the embeddings
    n_clusters = len(set(roles.values()))
    clustering_algorithm = KMeans(
        n_clusters=n_clusters,
        n_init=50,
        max_iter=500,
        random_state=0
    )
    clusters = find_clusters(np.array(servant_ids), model, clustering_algorithm, mode='e')

    # Prepare plotting dataframe
    plot_df = pd.DataFrame({
        "servant": map(names.get, servant_ids),
        "embedding1": embeddings_2d[:,0],
        "embedding2": embeddings_2d[:,1],
        "cluster": "cluster" + pd.Series(clusters).astype(str),
        "role": map(roles.get, servant_ids)
    }).dropna()

    # Create side-by-side plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
    fig.suptitle(f"Servant Clusters for Quest: {quest_name}")

    # Plot predicted clusters
    sns.scatterplot(data=plot_df, x="embedding1", y="embedding2",
                   hue="cluster", ax=ax1)
    ax1.set_title("Predicted Clusters")
    texts1 = [ax1.text(row['embedding1']+0.02, row['embedding2']+0.01,
                      str(row["servant"])) for _, row in plot_df.iterrows()]
    adjust_text(texts1, ax=ax1)

    # Plot true roles
    sns.scatterplot(data=plot_df, x="embedding1", y="embedding2",
                   hue="role", ax=ax2)
    ax2.set_title("True Roles")
    texts2 = [ax2.text(row['embedding1']+0.02, row['embedding2']+0.01,
                      str(row["servant"])) for _, row in plot_df.iterrows()]
    adjust_text(texts2, ax=ax2)

    plt.tight_layout()
    return plot_df


def plot_clusters(quest, link, color):
    draw_suggestion_clusters(
        graph,
        quest,
        link,
        model,
        extract_roles(get_suggestions(graph, quest, link), color),
        names
    )
```

### Classification

```python
skills = []
for skill_node in graph.subjects(RDF.type, FGO.Skill):
    id = int(next(graph.objects(skill_node, FGO.id)))
    skill = graph.qname(skill_node)

    priority_node = next(graph.objects(skill_node, FGO.hasPriority), None)
    priority = int(str(priority_node).split("_")[-1]) if priority_node else None

    owner = graph.qname(next(graph.subjects(FGO.hasSkill, skill_node)))

    if (FGO[owner], FGO.hasSkill1, skill_node) in graph:
        slot = 1
    elif (FGO[owner], FGO.hasSkill2, skill_node) in graph:
        slot = 2
    elif (FGO[owner], FGO.hasSkill3, skill_node) in graph:
        slot = 3

    skills.append((id, skill, owner, slot, priority))

skills_df = pd.DataFrame(skills, columns=["id", "name", "owner", "slot", "priority"])
skills_df
```

```python
train_skills_df = df[df["predicate"] == "hasPriority"]
valid_skills_df = df.drop(df.index)

s_train = skills_df[skills_df["name"].isin(train_skills_df["subject"])]
s_X_train, s_y_train = s_train.drop(columns=["priority"]), (s_train["priority"] + 2).astype(int)

s_valid = skills_df[skills_df["name"].isin(valid_skills_df["subject"])]
s_X_valid, s_y_valid = s_train.drop(columns=["priority"]), (s_valid["priority"] + 2).astype(int)

unseen = skills_df[skills_df["priority"].isna()]
unseen["prev_priority"] = unseen.apply(
    lambda row: skills_df.loc[
        (skills_df["owner"] == row["owner"]) &
        (skills_df["slot"] == row["slot"]) &
        skills_df["priority"].notna(),
        "priority"
    ].iloc[0] if not skills_df.loc[
        (skills_df["owner"] == row["owner"]) &
        (skills_df["slot"] == row["slot"]) &
        skills_df["priority"].notna(),
        "priority"
    ].empty else None,
    axis=1
)
unseen = unseen.dropna(subset=["prev_priority"])
unseen = unseen.drop(columns=["priority"])
unseen["prev_priority"] = (unseen["prev_priority"] + 2).astype(int)
unseen
```

### Link Prediction

```python
from typing import List

def get_suggestions(graph: Graph, quest_name: str, predicate: str) -> List[str]:
    """Get servant IDs suggested for a quest with given predicate."""
    FGO = Namespace("http://kg@se.ifmo.ru/~s367583/fgo#")
    return [graph.qname(obj) for obj in graph.objects(FGO[quest_name], FGO[predicate])]

suggestions = get_suggestions(graph, "christmas-2023-challenge-guide-2-hello-joulopukki", "hasSuggestedSupportServant")
for suggestion in suggestions:
    key = names.get(suggestion, "Unknown")
    print(key)
```

```python
from scipy.special import expit

merlin = inv_names["Merlin"]
koyanskaya = inv_names["Koyanskaya of Light"]

quests = []

for quest, _, _ in graph.triples((None, FGO.hasRecommendedSupportServant, FGO[merlin])):
    # koyanskaya also recommended
    if (quest, FGO.hasRecommendedSupportServant, FGO[koyanskaya]) not in graph:
        continue
    quests.append(quest)

servs = []
for i in graph.subjects(RDF.type, FGO.Servant):
  servs.append(i)
results = []
for quest in quests:
  # copy the servant list
  servlist = servs
  # removes the ones that ARE recommended
  for _, _, servant in graph.triples((quest, FGO.hasRecommendedSupportServant, None)):
    if servant in servlist:
      servlist.remove(servant)
  # add koyanskaya
  servlist.append(FGO[koyanskaya])
  statements = np.array(
    list({(graph.qname(quest), graph.qname(FGO.hasRecommendedSupportServant), graph.qname(i)) for i in servlist})
  );
  # calculate scores and probabilities
  scores = model.predict(statements)
  probs = expit(scores)
  probarray = probs.tolist()

  # koyanskaya has the last index
  koya = len(probarray)-1

  # sort the values with indexes attached to find koyanskaya's position later
  windex = sorted(enumerate(probarray), key=lambda x: x[1], reverse=True)
  wind = [index for index, _ in windex]
  # her new position
  new_index = wind.index(koya)
  results.append((graph.qname(quest), new_index, len(probarray), names[graph.qname(servlist[wind[0]])]))

results_df = pd.DataFrame(results, columns=["quest name", "koyanskaya's position", "total non-recommended", "winner servant"])
results_df
```

```python
from ampligraph.evaluation import train_test_split_no_unseen
import numpy as np

X_train, X_valid = train_test_split_no_unseen(np.array(df.to_numpy()), test_size=10000)
np.savez_compressed("data_splits.npz", X_train=X_train, X_valid=X_valid)
```

```python
import tensorflow as tf
from ampligraph.latent_features import ScoringBasedEmbeddingModel
from ampligraph.latent_features.loss_functions import get as get_loss
from ampligraph.latent_features.regularizers import get as get_regularizer

entities = set(X_train[:, 0]).union(set(X_train[:, 2]))
relations = set(X_train[:, 1])
max_ent_size = len(entities)
max_rel_size = len(relations)

model = ScoringBasedEmbeddingModel(k=100,
                                   eta=20,
                                   scoring_type='ComplEx',
                                   seed=0,
                                   max_ent_size=max_ent_size,
                                   max_rel_size=max_rel_size)

# Optimizer, loss and regularizer definition
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
loss = get_loss('multiclass_nll')
regularizer = get_regularizer('LP', {'p': 3, 'lambda': 1e-5})

# Compilation of the model
model.compile(optimizer=optimizer, loss=loss, entity_relation_regularizer=regularizer)
```

```python
model.fit(X_train,
          batch_size=int(X_train.shape[0] / 50),
          epochs=300, # Number of training epochs
          verbose=True # Displays a progress bar.
          )

```

```python
ranks = model.evaluate(X_valid,
                       batch_size=100,
                       use_filter={'train': X_train,
                                  'test': X_valid},
                       corrupt_side='s,o',
                       verbose=True)
from ampligraph.utils import save_model, restore_model
save_model(model, model_name_path="model.pkl")
```
