```python
from rdflib import Graph
```

```python
#load ontology
g = Graph()
g.parse(r"ontology\basic_ontlogy_characters.rdf")
```

# SPARQL queries

```python
def execute_query(query):
    results = g.query(query)
    return [(str(row[0]), str(row[1])) if len(row) > 1 else str(row[0]) for row in results]
```

```python
# Запрос 1: Список всех цивилизаций и их родных планет
query1 = """
SELECT ?civilization ?planet
WHERE {
  ?civilization a ex:civilization .
  ?civilization ex:bornIn ?planet .
}
"""
results1 = execute_query(query1)

# Запрос 2: Список всех персонажей и их видов (isSpecies)
query2 = """
SELECT ?character ?species
WHERE {
  ?character a ex:character .
  ?character ex:isSpecies ?species .
}
"""
results2 = execute_query(query2)

# Запрос 3: Персонажи, которые состоят в экипаже Normandy
query3 = """
SELECT ?character
WHERE {
  ?character a ex:character .
  ?character ex:inCrew ex:Normandy .
}
"""
results3 = execute_query(query3)

# Запрос 4: Персонажи с определённой способностью (Tactical_Armor)
query4 = """
SELECT ?character
WHERE {
  ?character a ex:character .
  ?character ex:hasAbility ex:Tactical_Armor .
}
"""
results4 = execute_query(query4)

# Запрос 5: Планеты и действия, которые на них можно выполнять
query5 = """
SELECT ?planet ?action
WHERE {
  ?planet a ex:Planet_Object .
  ?planet ex:hasAction ?action .
}
"""
results5 = execute_query(query5)

# Вывод результатов
results = {
    "Список всех цивилизаций и их родных планет (bornIn)": results1,
    "Список всех персонажей и их видов (isSpecies)": results2,
    "Персонажи, которые состоят в экипаже Normandy": results3,
    "Персонажи с определённой способностью (Tactical_Armor)": results4,
    "Планеты и действия, которые на них можно выполнять (hasAction)": results5
}

results
```
