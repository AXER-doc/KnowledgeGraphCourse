```python
from rdflib import Graph, Namespace, URIRef, Literal
from rdflib.namespace import RDF, RDFS, XSD
from owlready2 import *
import pandas as pd
```

```python
from rdflib import Namespace
EX = Namespace("http://www.semanticweb.org/sassh/ontologies/2024/11/untitled-ontology-5#")
g.bind("ex", EX)
```

```python
# load planets individs
csv_file = r"data\Mass Effect World Exploration.csv"
data = pd.read_csv(csv_file, sep=";")
data.fillna(method="ffill", inplace=True)

for _, row in data.iterrows():
    system = URIRef(EX[urllib.parse.quote(row["System"].replace(" ", "_"))])
    cluster = URIRef(EX[urllib.parse.quote(row["Cluster"].replace(" ", "_"))])
    planet = URIRef(EX[urllib.parse.quote(row["Planet/Object"].replace(" ", "_"))])
    action = URIRef(EX[urllib.parse.quote(row["Action"].replace(" ", "_"))])
    available = Literal(row["Available"] == 1)

    g.add((system, EX.hasCluster, cluster))
    g.add((cluster, EX.hasPlanet, planet))
    g.add((planet, EX.hasAction, action))
    g.add((action, EX.isAvailable, available))
    
    g.add((system, RDF.type, EX.System))
    g.add((cluster, RDF.type, EX.Cluster))
    g.add((planet, RDF.type, EX.Planet_Object))
    g.add((action, RDF.type, EX.Action))
```

```python
save_ontology_path =  r"ontology\basic_ontlogy_planets.rdf"
g.serialize(save_ontology_path, format="xml")
```

```python
## add characters
from rdflib import Graph
from rdflib import Namespace
g = Graph()
EX = Namespace("http://www.semanticweb.org/sassh/ontologies/2024/11/untitled-ontology-5#")
g.bind("ex", EX)
ontology_path = r"ontology\basic_ontlogy_civilizations.rdf"
g.parse(ontology_path, format="xml")
```

```python
data = pd.read_excel(r"data\me_characters.xlsx")
data
```

```python
ability_subclasses = {
    'Basic_Armor': 'Armor',
    'Combat_Armor': 'Armor',
    'Tactical_Armor': 'Armor',
    'Barrier': 'Biotic',
    'Lift': 'Biotic',
    'Singularity': 'Biotic',
    'Stasis': 'Biotic',
    'Throw': 'Biotic',
    'Warp': 'Biotic',
    'Assault_Training': 'Combat',
    'Fitness': 'Combat',
    'Assault_Rifles': 'Weapons',
    'Pistols': 'Weapons',
    'Shotguns': 'Weapons',
    'Sniper_Rifles': 'Weapons',
    'Unknown': 'Unknown'
}
```

```python
for _, row in data.iterrows():
    character = URIRef(EX[urllib.parse.quote(row['character'].replace(" ", "_"))])
    species = URIRef(EX[urllib.parse.quote(row['isSpecies'].replace(" ", "_"))])
    activity = URIRef(EX[urllib.parse.quote(row['doActivity'].replace(" ", "_"))])
    dna_type = URIRef(EX[row['hasDNA']])
    crew = URIRef(EX[row['inCrew']]) if row['inCrew'] != 'No' else None

    # Добавляем триплеты для персонажа
    g.add((character, RDF.type, EX.character))
    g.add((character, EX.isSpecies, species))
    g.add((character, EX.doActivity, activity))
    
    # Добавляем способности и их связь с подклассами abilities
    abilities = row['hasAbility'].split(',')
    if abilities != "Unknown":
        for ability in abilities:
            ability = ability.strip()
            ability_uri = URIRef(EX[urllib.parse.quote(ability)])
            subclass_uri = URIRef(EX[urllib.parse.quote(ability_subclasses[ability])])

            # Связываем способность с персонажем
            g.add((character, EX.hasAbility, ability_uri))

            # Указываем, что способность принадлежит к определенному подклассу abilities
            g.add((ability_uri, RDF.type, subclass_uri))
    
    # Добавляем ДНК
    g.add((character, EX.hasDNA, dna_type))
    
    # Если персонаж член экипажа, добавляем inCrew
    if crew:
        g.add((character, EX.inCrew, crew))
```

```python
save_ontology_path =  r"ontology\basic_ontlogy_characters.rdf"
g.serialize(save_ontology_path, format="xml")
```
