```python
ttl1 = 'urn_webprotege_ontology_9b013ff0-ebac-4953-a800-27a6cba6ed2e.ttl'
```

```python
ontology_path = folder_path+ttl1

# Создание графа
g = Graph()

# Загрузка онтологии в граф
g.parse(ontology_path)

# Сохранение графа в файл в формате Turtle
g.serialize(destination='ontology_output.ttl', format='turtle')

# Вывод графа в формате Turtle на экран
print(g.serialize(format='turtle'))
```

```python
ex = Namespace("http://example.org/")
g.bind("ex", ex)

# Добавление данных в онтологию

# Примеры мемов
Statham = URIRef(ex.Statham)
Povar = URIRef(ex.Povar)

# Примеры эмоции
Motivation = URIRef(ex.Motivation)
Joy = URIRef(ex.Joy)
```

```python
# Добавление триплетов
g.add((Motivation, RDF.type, ex.Emotion))
g.add((Statham, RDF.type, ex.Meme))
g.add((Statham, ex.expressEmotion, Motivation))


g.add((Joy, RDF.type, ex.Emotion))
g.add((Povar, RDF.type, ex.Meme))
g.add((Povar, ex.expressEmotion, Joy))
```

```python
g.serialize(destination="meme_ontology.rdf", format="turtle")

# Вывод графа на экран
print(g.serialize(format="turtle"))
```

```python
# Запрос SPARQL для получения количества эмоций
query = """
PREFIX ex: <http://example.org/>
PREFIX wp: <http://webprotege.stanford.edu/>

SELECT (COUNT(DISTINCT ?Emotion) AS ?count)
WHERE {
  ?Emotion a ex:Emotion .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(row)
    print(f"Количество эмоций: {row['count']}")
```

```python
# Запрос SPARQL для получения количества мемов с мотивацией
query = """
PREFIX ex: <http://example.org/>

SELECT (COUNT(DISTINCT ?Meme) AS ?count)
WHERE {
  ?Meme a ex:Meme ;
        ex:expressEmotion ex:Motivation .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(f"Количество мемов с мотивацией: {row['count']}")
```
